## Optimized structure for $\lambda_t=5.0\mu m$   

```python
# list of layer thicknesses in um
d_list = [inf,0.42,0.63,0.42,1.05,0.63,0.63,1,inf]
# list of refractive indices
n_list = [1,4.182,1.3475,4.182,1.3475,4.182,1.3475,2.6673+22.612j,1]
# list of wavenumbers to plot in um^-1
ks=linspace(4.00,8.00,num=400)
# spectral directional emissivity = 1-R_lambda 
emissivity = []
# initialize lists of y-values to plot
for k in ks:
    emissivity.append(1-coh_tmm('p', n_list, d_list, 0, k)['R'])
# plot figure
plt.figure()
plt.ylabel('Spectral directional emissivity')
plt.xlabel('Wavelength ($\mu$$m^{-1}$)')
plt.plot(ks,emissivity,'red')
```

## Optimized structure for $\lambda_t=5.0\mu m$,$\lambda_t=6.0\mu m$ and $\lambda_t=7.0\mu m$  

```python
# struture for lambda_t = 5um
d_list_5 = [inf,0.42,0.63,0.42,1.05,0.63,0.63,1,inf]
n_list_5 = [1,4.182,1.3475,4.182,1.3475,4.182,1.3475,2.6673+22.612j,1]
# struture for lambda_t = 6um
d_list_6 = [inf,0.42,0.63,0.42,0.85,0.85,0.63,1,inf]
n_list_6 = [1,4.1747,1.2713,4.1747,1.2713,4.1747,1.2713,3.7111+27.316j,1]
# struture for lambda_t = 7um
d_list_7 = [inf,0.44,0.66,0.44,0.88,0.44,0.22,0.44,0.44,1,inf]
n_list_7 = [1,4.1703,1.1145,4.1703,1.1145,4.1703,1.1145,4.1703,1.1145,4.9405+31.765j,1]
# list of wavenumbers to plot in um^-1
ks=linspace(4.00,8.00,num=400)
# spectral directional emissivity = 1-R_lambda 
emissivity_5 = []
emissivity_6 = []
emissivity_7 = []
# initialize lists of y-values to plot
for k in ks:
    emissivity_5.append(1-coh_tmm('p', n_list_5, d_list_5, 0, k)['R'])
    emissivity_6.append(1-coh_tmm('p', n_list_6, d_list_6, 0, k)['R'])
    emissivity_7.append(1-coh_tmm('p', n_list_7, d_list_7, 0, k)['R'])
# plot figure
plt.figure()
plt.ylabel('Spectral directional emissivity')
plt.xlabel('Wavelength ($\mu$$m^{-1}$)')
plt.plot(ks,emissivity_5,'red',ks,emissivity_6,'blue',ks,emissivity_7,'green')
```
